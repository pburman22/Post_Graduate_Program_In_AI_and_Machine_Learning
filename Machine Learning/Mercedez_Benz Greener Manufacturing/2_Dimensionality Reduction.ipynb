# Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

# Import Datasets

```python
# training data
train_raw_pre = pd.read_csv('Dataset/Pre Processed Data/train_raw_pre.csv')

# testing data
test_raw_pre = pd.read_csv('Dataset/Pre Processed Data/test_raw_pre.csv')
```

# Principal Component Analysis (PCA)

```python
# view the train data
train_raw_pre.head()
```

```python
# view the test data
test_raw_pre.head()
```

```python
# drop the columns 'Unnamed: 0'
train_raw_pre = train_raw_pre.drop(['Unnamed: 0'], axis = 1)
test_raw_pre = test_raw_pre.drop(['Unnamed: 0'], axis = 1)
```

```python
train_raw_pre.head()
```

```python
test_raw_pre.head()
```

```python
# Loading the PCA object for training set
pca = PCA(n_components = 2) # Here we choose the number of components that we will keep.
pca_train = pca.fit_transform(train_raw_pre)
pca_train_df = pd.DataFrame(pca_train, columns = ['principal_component_1','principal_component_2'])

# Loading the PCA object for test set
pca = PCA(n_components = 2) # Here we choose the number of components that we will keep.
pca_test = pca.fit_transform(test_raw_pre)
pca_test_df = pd.DataFrame(pca_test, columns = ['principal_component_1','principal_component_2'])
```

```python
# view the train set
pca_train_df.info()
```

```python
# view the test set
pca_test_df.info()
```

**Export the dataset**

```python
# export training set
pca_train_df.to_csv('Dataset/PCA Data/pca_train.csv')

# export testing set
pca_test_df.to_csv('Dataset/PCA Data/pca_test.csv')
```
